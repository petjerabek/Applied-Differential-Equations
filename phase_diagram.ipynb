```python
import numpy as np
import matplotlib.pyplot as plt

x_vals = np.linspace(-5, 5, 25)
y_vals = np.linspace(-5, 5, 25)
X, Y = np.meshgrid(x_vals, y_vals)

X_prime = Y**2 - X
Y_prime = X*Y + X

N = np.sqrt(X_prime**2 + Y_prime**2)
N[N == 0] = 1 # prevent division by zero at equilibrium points
X_prime_norm = X_prime / N
Y_prime_norm = Y_prime / N

plt.figure(figsize=(10, 8))

plt.quiver(X, Y, X_prime_norm, Y_prime_norm, color='gray', alpha=0.7)

y_nc = np.linspace(-5, 5, 100)
x_nc = y_nc**2
plt.plot(x_nc, y_nc, 'r-', linewidth=2.5, label="$x'=0$ ($x=y^2$)")

plt.axvline(x=0, color='b', linestyle='--', linewidth=2.5, label="$y'=0$ ($x=0$)")
plt.axhline(y=-1, color='b', linestyle='--', linewidth=2.5, label="$y'=0$ ($y=-1$)")

plt.plot(0, 0, 'ko', label='eq. point (0,0)')
plt.plot(1, -1, 'ko', label='eq. point(1,-1)')

plt.xlim(-5, 5)
plt.ylim(-5, 5)

plt.title("Phase portrait ($x' = y^2 - x$ and $y' = xy + x$)")
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.8) # x-axis
plt.legend()

plt.show()
```
